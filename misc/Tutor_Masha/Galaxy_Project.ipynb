```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from random import gauss
from random import random
from random import uniform
from random import choice
```

```python
def rad_prob(r,c,lim):
    r=abs(r)
    return lim/((r/c+lim)*(1+r/c)**2)
r=np.linspace(-100,100,10000)
p=rad_prob(r,20,0.01)
plt.plot(r,p)
plt.show()

def gaussian(x,mu,sigma):
    return np.exp(-4*np.log(2)*((x-mu)**2)/sigma**2)

def spiral(theta,shift,r,posi,inten,strt,leng):
    x=0.075
    for i in range(len(posi)):
        x+=(inten[i])*gaussian(theta+shift,posi[i],0.2*gaussian(r,strt[i],100*leng[i]))
    return x
```

```python
def gen_rads(n,ran,c,lim):
    rads=[]
    for i in range(n):
        r=uniform(0,ran)
        prob=rad_prob(r,c,lim)
        while random()>prob:
            r=uniform(0,ran)
            prob=rad_prob(r,c,lim)
        rads.append(r)
    return np.array(rads)

def gen_uni_angl():
    angl=[]
    for i in range(len(rads)):
        angl.append(uniform(0,2*np.pi))
    return angl

def gen_spi_angl(rads,ran,spi_deg,n):
    angl=[]
    position=[]
    intnsty=[]
    strt=[]
    length=[]
    for i in range(n):
        position.append(random()*np.pi+np.pi)
        intnsty.append(0.95*random()+0.05)
        strt.append(9*ran*random()/10+ran/10)
        length.append(random())
    for i in range(len(rads)):
        r=rads[i]
        shift=r*spi_deg
        theta=uniform(0,2*np.pi)+shift
        while random()>spiral(theta,shift,r,position,intnsty,strt,length):
            theta=uniform(0,2*np.pi)+shift
        
        angl.append(theta+choice([0,1])*np.pi)
    return angl

def to_xy(rads,angl):
    x=[]
    y=[]
    for i in range(len(rads)):
        x.append(rads[i]*np.cos(angl[i]))
        y.append(rads[i]*np.sin(angl[i]))
    return x,y
```

##Uniform

```python
ran=400.
rads=gen_rads(40000,ran,100.,10.)
angl=gen_uni_angl()

x,y=to_xy(rads,angl)
plt.figure(figsize=(10,10))
plt.scatter(x,y,marker='.',linewidths=0,color='k',s=2)
plt.show()
```

##Spiral

```python
ran=400.
rads=gen_rads(10000,ran,800.,0.1)
angl=gen_spi_angl(rads,ran,0.01,20)

x,y=to_xy(rads,angl)
plt.figure(figsize=(12,12))
plt.scatter(x,y,marker='.',linewidths=0,color='k',s=2)
plt.show()
```
