```python
def rad_prob(r,c,lim):
    r=abs(r)
    return lim/((r/c+lim)*(1+r/c)**2)
r=np.linspace(-100,100,10000)
p=rad_prob(r,0.5,0.1)
plt.plot(r,p)
plt.show()

def gaussian(x,mu,sigma):
    return np.exp(-4*np.log(2)*((x-mu)**2)/sigma**2)

def spiral(theta,shift,r,posi,inten,strt,leng):
    x=0.075
    for i in range(len(posi)):
        x+=(inten[i])*gaussian(theta+shift,posi[i],0.2*gaussian(r,strt[i],100*leng[i]))
    return x
```

```python
def gen_rads(n,ran,c,lim):
    rads=[]
    for i in range(n):
        r=uniform(0,ran)
        prob=rad_prob(r,c,lim)
        while random()>prob:
            r=uniform(0,ran)
            prob=rad_prob(r,c,lim)
        rads.append(r)
    return np.array(rads)

def gen_uni_angl():
    angl=[]
    for i in range(len(rads)):
        angl.append(uniform(0,2*np.pi))
    return angl

def gen_spi_angl(rads,ran,spi_deg,n):
    angl=[]
    position=[]
    intnsty=[]
    strt=[]
    length=[]
    for i in range(n):
        position.append(random()*np.pi+np.pi)
        intnsty.append(random())
        strt.append((ran*3.9*random()+1.15*ran)/5)
        length.append(random())
    for i in range(len(rads)):
        r=rads[i]
        shift=r*spi_deg
        theta=uniform(0,2*np.pi)+shift
        while random()>spiral(theta,shift,r,position,intnsty,strt,length):
            theta=uniform(0,2*np.pi)+shift
        
        angl.append(theta+choice([0,1])*np.pi)
    return angl

def to_xy(rads,angl):
    x=[]
    y=[]
    for i in range(len(rads)):
        x.append(rads[i]*np.cos(angl[i]))
        y.append(rads[i]*np.sin(angl[i]))
    return x,y
```

##Uniform

```python
ran=50.
rads=gen_rads(500,ran,5,0.1)
angl=gen_uni_angl()

x,y=to_xy(rads,angl)
plt.figure(figsize=(10,10))
plt.scatter(x,y,marker='.',linewidths=0,color='k',s=100)
plt.show()
```

```python
def gen_vel(r,Ct1,Ct2,Cr):
    tdot=np.array([gauss(Ct1,Ct2)/np.sqrt(i) for i in r]) #(1-np.e**(-(i/gauss(10,1))**1.1))
    rdot=np.array([gauss(0,Cr) for i in r])
    return tdot,rdot

tdot,rdot=gen_vel(rads,0.2,0.05,1)

plt.figure(figsize=(10,5))
plt.scatter(rads,tdot,linewidths=0,s=3)    
```

```python
def to_xydot(r,t,tdot,rdot):
    '''returns velocities in x-y coordinates'''
    xdot=rdot*np.cos(t)-r*np.sin(t)*tdot
    ydot=rdot*np.sin(t)+r*np.cos(t)*tdot
    return xdot,ydot
```

```python
xdot,ydot=to_xydot(rads,angl,tdot,0)
plt.figure(figsize=(10,10))
plt.scatter(xdot,ydot,linewidths=0,s=2)
plt.show()
```

```python
def move(x,y,xdot,ydot):
    x=x+xdot
    y=y+ydot
    return x,y
def accel(S,M,x,y,xdot,ydot,Sx,Sy):
    r=np.sqrt(x**2+y**2)
    R=np.sqrt(Sx**2+Sy**2)
    xacc=6.67e-11*x/r*(-M/x**3+S/(R-r)**3+S/R**3)
    yacc=6.67e-11*y/r*(-M/y**3+S/(R-r)**3+S/R**3)
    return xdot-xacc,ydot-yacc
```

```python
xplot=[]
yplot=[]
xdotplot=[]
ydotplot=[]
for i in range(20):
    xplot.append(x)
    yplot.append(y)
    xdotplot.append(xdot)
    ydotplot.append(ydot)
    x,y=move(x,y,xdot,ydot)
    xdot,ydot=accel(1e20,0,x,y,xdot,ydot,1,1)
    plt.figure(figsize=(10,10))
    plt.scatter(xplot,yplot,linewidths=0,s=1,color='k')
    plt.xlim(-500,500)
    plt.ylim(-500,500)
    plt.show()
#plt.figure(figsize=(10,10))
#plt.scatter(xplot,yplot,linewidths=0,s=1,color='k')
#plt.xlim(-1000,1000)
#plt.ylim(-1000,1000)
#plt.show()
```
