```python
def is_leap(y):
    '''determines if it's a leap year and returns True of False'''
    leap=(y%4==0,y%400!=0)
    true=(True,True)
    if leap==true:
        return True
    if leap!=true:
        return False

def d_in_m(m,l):
    '''finds the number of days in a month given the year and the month'''
    if l==True:
        if m==1:
            m=m-1
        if m==2:
            m=2
        return month[m]
    if l==False:
        if m==1:
            m=m-1
        if m==2:
            m=1
        return month[m]
```

```python
count=0
meter=1
for y in range(1900,2000):
    l=is_leap(y)
    for m in range(1,13):
        meter+=d_in_m(m,l)%28
        if meter>6:
            meter=meter-7
        if meter==0:
            count+=1
print count
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt

k=8.98e9
fig = plt.figure(figsize=(15,10))
ax = fig.gca(projection='3d') #gca stands for get current axis.
X = np.linspace(-2.3,2.3,100)
Y = np.linspace(-2.3,2.3,100)
X, Y = np.meshgrid(X, Y)

def R(x,y,locx,locy):
    return np.sqrt((x-locx)**2+(y-locy)**2)
Z=-k/R(X,Y,1,0)-k/R(X,Y,-1,0)+k/R(X,Y,0,1)+k/R(X,Y,0,-1)

surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.seismic, linewidth=0.07, shade=True)

fig.colorbar(surf, shrink=0.5)

plt.show()
```
