```python
def two_gaus_tool(shift_min,shift_max,num_graph,plot_min,plot_max,num_points,g1_amp,g2_amp,g1_sig,g2_sig,g1_sep_rat=1,g2_sep_rat=1,plot_gaus=True,graph_frac=None):
    '''GENERAL
    -------
    plots two gaussians displaying a graph of the summed gaussians with a plot of the apparent left and right maximums overlaying it, an isolated graph of the left and right maximums, and a plot of the true seperation (distance between original maximums) against the difference between the true seperation and the apparent seperation (the apparent seperation between maximums). 
    
    FIELD INFO
    ----------
    shift_min/shift_max:
    how much each graph will shift left or right (shifting is symetrical about x=0).
    
    num_graph:
    the number of double gaussians which will be analyzed within the shifting limits plot_min/plot_max: range of x values for which the gaussians are calculated
    
    num_points:
    number of x values within the range for which the gaussians are calculated (analagous to plotting accuracy).
    
    plot_gaus:
    if apparrent and true seperations data is all that is required leave field empty.
    
    graph_frac:
    if plot_gauss=True graph frac determines what fraction of the graphs which are used in calculations will actually be plotted. default is to plot 10 graphs
    
    GAUSSIAN CHARACTERISTICS
    ------------------------
    g1_amp/g2_amp:
    amplitude of the two gausians (issues still being worked out for g1_amp!=g2_amp case).
    
    g1_sig/g2_sig:
    sigma values for the two gaussians (issues still being worked out for g1_sig!=g2_sig case).
    
    g1_sep_rat/g2_sep_rat: define the rate of seperation for each gaussian (default value = 1).'''
    
    if g1_sep_rat<0:
        if abs(g1_sep_rat*shift_max)>=abs(plot_max):
            raise ValueError('sep_rat must be less than the quotient of (respective plot limit)/(shift_max); increase plot min or max.')
    if g1_sep_rat>0:
        if abs(g1_sep_rat*shift_max)>=abs(plot_min):
            raise ValueError('sep_rat must be less than the quotient of (respective plot limit)/(shift_max); increase plot min or max.')

    if g2_sep_rat<0:
        if abs(g2_sep_rat*shift_max)>=abs(plot_min):
            raise ValueError('sep_rat must be less than the quotient of (respective plot limit)/(shift_max); increase plot min or max.')
    if g2_sep_rat>0:
        if abs(g2_sep_rat*shift_max)>=abs(plot_max):
            raise ValueError('sep_rat must be less than the quotient of (respective plot limit)/(shift_max); increase plot min or max.')
    
    
    if graph_frac==None:
        graph_frac=num_graph/10
    
    def gaus(x,a,b,c):
        return a*e**(-float((x-b)**2)/2/c**2)
    
    max_range=linspace(shift_min,shift_max,num_graph,)
    figsize(10,8)
    count=graph_frac
    x_sep_t=[]
    x_ta_dif=[]
    max_dist=[]
    l_x=[]
    l_y=[]
    r_x=[]
    r_y=[]
    for i in max_range:
        x=linspace(plot_min,plot_max,num_points)
        y=[]
        y1=[]
        y2=[]
        for j in x:
            y.append(gaus(j,g1_amp,-i*g1_sep_rat,g1_sig)+gaus(j,g2_amp,i*g2_sep_rat,g2_sig))
        if plot_gaus==True:
            if count%graph_frac==0:
                plot(x,y,color='k',linestyle='--',linewidth=.5)
        ysort=sort(y)
        x=array(x)
        y=array(y)
        maxs=y[1:-1][(diff(y)[0:-1]>0)&(diff(y)[1:]<0)]
        
        if len(maxs)==0:
            maxs=([max(y)])

        max_i=[]
        for k in maxs:
            for l in sort(where(y==k)[0]): max_i.append(l)

        if len(max_i)>0:
            
            if plot_gaus==True:
                if len(max_i)!=1:
                    l_x.append(x[max_i[1]])
                    l_y.append(y[max_i[1]])
                    r_x.append(x[max_i[0]])
                    r_y.append(y[max_i[0]])
                if len(max_i)==1:
                    l_x.append(x[max_i[0]])
                    l_y.append(y[max_i[0]])
                    r_x.append(x[max_i[0]])
                    r_y.append(y[max_i[0]])
                    
            if len(max_i)<=1:
                x_ta_dif.append(2*i)
            if len(max_i)!=1:
                x_ta_dif.append((abs(g1_sep_rat*i)+abs(g2_sep_rat*i)-abs((x[max_i[0]]-x[max_i[1]]))))
            x_sep_t.append(2*i)
            
        if count==graph_frac:
            amp_dif=g1_amp-g2_amp
            if amp_dif<=0:
                yplotlim=2*g2_amp+0.025*g2_amp
            if amp_dif>0:
                yplotlim=2*g1_amp+0.025*g1_amp
        count+=1
        
        
        
    if plot_gaus==True:
        plot(l_x,l_y,color='c',linewidth=1.75,label='right max')
        plot(r_x,r_y,color='r',linestyle='--',linewidth=2,label='left max')
        ylabel('Y',fontsize=14)
        xlabel('X',fontsize=14)
        xlim(plot_min,plot_max)
        ylim(0,yplotlim+0.09*yplotlim)
        legend()
        show()
        
        figsize(6,5)
        plot(l_x,l_y,color='c',label='right max',linewidth=1.75)
        plot(r_x,r_y,color='r',label='left max',linestyle='--',linewidth=2)
        dif=g1_amp-g2_amp
        xlim(-shift_max,shift_max)
        if dif<=0:
            ylim(g1_amp,yplotlim+0.05*yplotlim)
        else:
            ylim(g2_amp,yplotlim+0.05*yplotlim)
        ylabel('Y',fontsize=14)
        xlabel('X',fontsize=14)
        legend()
        show()
        
        figsize(8,7)
        plot(x_sep_t,x_ta_dif,color='b')
        ylabel('Difference Between Apparent and True Seperation',fontsize=14)
        xlabel('True Seperation',fontsize=14)
        xlim(shift_min*2,shift_max*2)
        ylim_max=max(x_ta_dif)
        ylim(-0.02*ylim_max,ylim_max+0.05*ylim_max)
        show()
```
