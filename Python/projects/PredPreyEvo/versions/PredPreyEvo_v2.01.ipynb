```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random as rand
%matplotlib inline
```

```python
class org(object):
    def __init__(self,key,mut_var,rep_pnt,rep_don):
        '''Initialize an organism with plant-like atributes
        key: list of numbers which ultimately determine an organism's fitness
        mut_var: The mutation rate for the organism.
        rep_don: The nutrients donated to offspring during preproduction
        rep_pnt: An organism reproduces when fat stores exceded rep_pnt
        '''
        self.mut_var = mut_var + rand.gauss(0,mut_var)*np.log(1+mut_var)
        self.rep_don = rep_don + rand.gauss(0,mut_var)*np.log(1+rep_don)
        self.rep_pnt = rep_pnt + rand.gauss(0,mut_var)*np.log(1+rep_pnt)
        # Initialize Organism's Key
        # ------------------------
        # The key serves multiple purposes and is the central
        # fitness measure. The key is compared to other keys
        # from the enviroment or other organisms. The difference
        # between the individual values of the compared keys is
        # what determines how much energy is transfered between
        # the entities in volved.
        self.key = [key[i]+rand.gauss(0,mut_var) for i in range(len(key))]
        self.key += key[int(round(len(key)-1-rand.gauss(0,mut_var)*np.exp(10-len(key)))):len(key)-1]
        # Initialize Organism's Fat
        # -------------------------
        # Fat is equivalent to the amount of nutrients the organism
        # has stored up. If an organisms fat drops to, or below
        # zero then it dies. If the fat stores exceded rep_pnt (the
        # reproduction point) then the organism reproduces and
        # passes on its characteristics along with the resulting
        # mutations that come along with the process.
        self.fat = rep_don + rand.gauss(0,mut_var)*rep_don
    
    def get_id(self):
        return hex(id(self))
    
    def subsist(self):
        '''
        Returns
        -------
        False: if fat<0 (indicates that the organism is dead)
        True: if 0<fat<rep_pnt (indicates that the organism has survived)
        '''
        self.fat -= len(self.key)*0.07
        if self.fat<0:
            return False
        else:
            return True
        
    def persist(self):
        '''Performs subsist() and if fat>rep_pnt, reproduce()
        
        Returns
        -------
        subsist: True if organism is alive and False if it's dead
        offspring: None if fat<rep_pnt, otherwise returns reproduce()
        '''
        if self.fat>self.rep_pnt:
            offspring = self.reproduce()
        else:
            offspring = None
        return self.subsist(),offspring
```

```python
class plnt_org(org):
    def reproduce(self):
        '''Returns a new spawn with inhereted traits from self'''
        self.fat -= self.rep_don
        return plnt_org(self.key,self.mut_var,self.rep_don,self.rep_pnt)
        
    def consume(self,nutrient_key):
        '''The organism will gather nutrients based on the given nutrient key
        
        nutrient_key: a list of values representing encoded energy from the system
        '''
        if nutrient_key!=None:
            key_slc = min([len(nutrient_key),len(self.key)])
            nutr_key = nutrient_key[:key_slc]
            self_key = self.key[:key_slc]
            diff = sum([abs(nutr_key[i]-self_key[i]) for i in range(key_slc)])
            self.fat += 1./(diff+0.01)
```

```python
class anml_org(org):
    def __init__(self,key,mut_var,rep_pnt,rep_don,hrm_pnt):
        '''Initialize an organism with animal-like attributes
        
        key: list of numbers which ultimately determine an organism's fitness
        mut_var: The mutation rate for the organism.
        rep_don: The nutrients donated to offspring during preproduction
        rep_pnt: An organism reproduces when fat stores exceded rep_pnt
        hrm_pnt: The point at which poor predation results are harmful
        '''
        super(anml_org,self).__init__(key,mut_var,rep_pnt,rep_don)
        self.hrm_pnt = hrm_pnt + rand.gauss(0,mut_var)*hrm_pnt
        self.pred_mem={'plnt_org':0,'anml_org':0}
    
    def reproduce(self):
        '''Returns a new spawn with inhereted traits from self'''
        self.fat -= self.rep_don
        return anml_org(self.key,self.mut_var,self.rep_don,self.rep_pnt,self.hrm_pnt)
    
    def predpref(self,name):
        if self.pred_mem['plnt_org']+self.pred_mem['anml_org']==0:
            return 0.5
        else:
            return self.pred_mem[name]/(self.pred_mem['plnt_org']+self.pred_mem['anml_org'])
    
    def predate(self,group):
        '''The organism attemps to steal nutrients from its victim
        
        group: a population of plant_org and anml_org given as a list of objects
        
        The case without a predation attempt:
        >> If the group size is too small there is a chance that 
        
        The case for a failed predation attempt:
        >> There is no failure state for a predation attemp on a plant_org.
        >> However, if vict is an anml_org, this causes the predator to
           lose fat for having attempted, but failed to catch anything.
        
        The case for a succesful predation attempt:
        >> The vict.key and self.key are sufficiently similar and the
           predator steals nutrients from its victim.
        >> The vict.key and self.key are dissimilar such that the sum of
           their differences is greater than hrm_pnt. In this case the
           nutrients from the predator is not able to digest the victim's
           nutrients. This causes both the predator and the victim to
           lose fat; the victim for having been damaged and the predator
           for having been, in a sense, poisoned.
           
        '''
        if 5*np.log(1-rand.random())/-len(group.members)**.25>0.5:
            break_check = False
            vict = group.members[rand.randrange(0,len(group.members)-1)]
            for i in range(int(self.fat*2)):
                if rand.random()>self.predpref(vict.__class__.__name__):
                    vict = group.members[rand.randrange(0,len(group.members)-1)]
                else:
                    self.pred_mem[vict.__class__.__name__]+=1
                    break_check = True
                    break
            if break_check==False:
                vict = None
        else:
            vict = None
        f1 = self.fat*1.0
        if vict!=None:
            if str(vict.__class__) == str(anml_org):
                # Make a comparison of fat levels to determine the
                # predator's and the victim's health. If more fat
                # is attributed to the predator it's more likely that
                # it will steal fat from its victim or vice versa.
                if rand.random()<self.fat/(self.fat+vict.fat):
                    key_slc = min([len(self.key),len(vict.key)])
                    # In predation of anml_orgs the begining of
                    # the victim's key is analyzed.
                    vict_key = vict.key[-key_slc:]
                    pred_key = self.key[-key_slc:]
                    diff = sum([abs(pred_key[i]-vict_key[i]) for i in range(key_slc)])
                    # Inclusion of a hrm_pnt means the predator can be harmed if
                    # its key and the victim's are too dissimilar.
                    diff = diff*(self.hrm_pnt-diff)/abs(self.hrm_pnt-diff)/self.hrm_pnt
                    fat_cut = len(self.key)/(diff+0.005)/3
                    # Make sure that the predator only gets the amount of fat
                    # which the victim contains, rather than how much it could
                    # steal.
                    if fat_cut>vict.fat:
                        self.fat += abs(vict.fat)
                        vict.fat -= fat_cut
                    else:
                        self.fat += fat_cut
                        vict.fat -= abs(fat_cut)
                else:
                    # If the predation attempt is unsuccesful then the victim
                    # gets away with less harm while the predator wastes more fat
                    # making the attempt and failing.
                    self.fat -= len(self.key)*0.2
                    vict.fat -= vict.fat*0.05
                
                    
            if str(vict.__class__) == str(plnt_org):
                # There is no failure state for predation attempts
                # on plnt_org types.
                key_slc = min([len(self.key),len(vict.key)])
                vict_key = vict.key[:key_slc]
                pred_key = self.key[:key_slc]
                diff = sum([abs(pred_key[i]-vict_key[i]) for i in range(key_slc)])
                diff = diff*(self.hrm_pnt-diff)/abs(self.hrm_pnt-diff)/self.hrm_pnt
                fat_cut = len(self.key)/(diff+0.005)/15
                if fat_cut>vict.fat:
                    self.fat += vict.fat
                    vict.fat -= fat_cut
                else:
                    self.fat += fat_cut
                    vict.fat -= abs(fat_cut)
```

```python
class group:
    def __init__(self):
        self.members = []
        
    def gen_pop(self,size,ratio,plnt_args,anml_args):
        '''Initialize a population of plant and animal type organisms
        
        gen_pop Parameters
        ------------------
        size: the number of organisms in the population
        ratio: the percent of the population composed of plants
        plnt_args: arguments for initializing organisms of type plnt_org
        anml_args: arguments for initializing organisms of type anml_org
        plnt_var: the variance of the parameters given in plnt_args
        anml_var: the variance of the parameters given in anml_args
        
        plnt_orgs Parameters
        --------------------
        key: list of numbers which ultimately determine an organism's fitness
        mut_var: The mutation rate for the organism.
        rep_pnt: An organism reproduces when fat stores exceded rep_pnt
        rep_don: The nutrients donated to offspring during preproduction
        
        anml_orgs Parameters
        --------------------
        key: list of numbers which ultimately determine an organism's fitness
        mut_var = The mutation rate for the organism.
        rep_pnt = An organism reproduces when fat stores exceded rep_pnt
        rep_don = The nutrients donated to offspring during preproduction
        hrm_pnt = The point at which poor predation results are harmful
        '''
        self.members = [plnt_org(*plnt_args) if rand.random()<ratio else anml_org(*anml_args) for i in range(size)]
        
    def remove(self,org_id):
        '''Remove the given organism from self.members
        
        org_id: a hex string identifying the organism's memory address'''
        for i in range(len(self.members)):
            if self.members[i].get_id()==org_id:
                self.members.pop(i)
                return
        raise KeyError('hex string not found in self.members')
    
    def remove_set(self,start,end):
        '''Remove self.members[sart:end] from self.memebrs where end is exclusive'''
        self.members = self.members[:start] + self.members[end:]
        return
        
    def append(self,org):
        '''Add the given organism to self.members'''
        self.members.append(org)
        return
    
    def append_set(self,new_members):
        '''Add a set of organisms to a preexisting or empty population'''
        self.members += new_members
        return
    
    def get_ratio(self):
        '''Returns the percent of the population composed of plnt_org types'''
        plnt_count = 0
        for org in self.members:
            if str(org.__class__) == str(plnt_org):
                plnt_count+=1.
        return plnt_count/len(self.members),plnt_count,len(self.members)-plnt_count
    
    def get_traits(self):
        '''Get two lists with the unique attributes of plant and animal type organisms
        
        Returns
        -------
        A dict object with the following keys:
        plnt_traits = a list tuples holding the unique traits of the population's plants
        anml_traits = a list tuples holding the unique traits of the population's animals
        '''
        plnt_traits = {'mut_var':[],'rep_don':[],'rep_pnt':[],'fat':[]}
        anml_traits = {'mut_var':[],'rep_don':[],'rep_pnt':[],'hrm_pnt':[],'fat':[]}
        for org in self.members:
            if str(org.__class__) == str(plnt_org):
                plnt_traits['mut_var'].append(org.mut_var)
                plnt_traits['rep_don'].append(org.rep_don)
                plnt_traits['rep_pnt'].append(org.rep_pnt)
                plnt_traits['fat'].append(org.fat)
            if str(org.__class__) == str(anml_org):
                anml_traits['mut_var'].append(org.mut_var)
                anml_traits['rep_don'].append(org.rep_don)
                anml_traits['rep_pnt'].append(org.rep_pnt)
                anml_traits['hrm_pnt'].append(org.hrm_pnt)
                anml_traits['fat'].append(org.fat)
        return {'plnt_org':plnt_traits,'anml_org':anml_traits}
    
    def get_keys(self):
        '''Get two lists with the unique keys of plant and animal type organisms
        
        Returns
        -------
        plnt_keys: a list holding the keys of the population's plants
        anml_keys: a list holding the keys of the population's animals
        '''
        plnt_keys = []
        anml_keys = []
        for org in self.members:
            if str(org.__class__) == str(plnt_org):
                plnt_keys.append(org.key)
            if str(org.__class__) == str(anml_org):
                anml_keys.append(org.key)
        return plnt_keys,anml_keys
```
