```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random as rand
%matplotlib inline
```

```python
class phy_organism:
    def __init__(self,mut_var,rep_don,rep_pnt):
        '''
        mut_var: The mutation rate for the organism.
        rep_don: The nutrients donated to offspring during preproduction
        rep_pnt: An organism reproduces when fat stores exceded rep_pnt
        '''
        self.mut_var = mut_var
        self.rep_don = rep_don
        self.rep_pnt = rep_pnt
        # Initialize Organism's Key
        # ------------------------
        # The key serves multiple purposes and is the central
        # fitness measure. The key is compared to other keys
        # from the enviroment or other organisms. The difference
        # between the individual values of the compared keys is
        # what determines how much energy is transfered between
        # the entities in volved.
        self.key = [rand.random() for i in range(1+int(rand.random()*9))]
        # Initialize Organism's Fat
        # -------------------------
        # Fat is equivalent to the amount of nutrients the organism
        # has stored up. If an organisms fat drops to, or below
        # zero then it dies. If the fat stores exceded rep_pnt (the
        # reproduction point) then the organism reproduces and
        # passes on its characteristics along with the resulting
        # mutations that come along with the process.
        self.fat = rep_don
    
    def sustain(self,nutrient_key=None):
        '''
        Returns
        -------
        offspring: a spawn with inhereted traits from self if fat>rep_pnt
        False: if fat<0 (indicates that the organism is dead)
        True: if 0<fat<rep_pnt (indicate the organism has sustained itself)
        '''
        self.fat -= len(self.key)/2
        if self.fat>self.rep_pnt:
            self.fat -= self.rep_don
            key_len = int(round(len(self.key)+rand.gauss(0,self.mut_var)))
            new_key = [self.key[i]+rand.gauss(0,self.mut_var) if i>len(self.key) else rand.random() for i in range(key_len)]
            new_mutvar = self.mut_var+rand.gauss(0,self.mut_var/2)
            new_reppnt = self.rep_pnt+rand.gauss(0,self.mut_var*4)
            offspring = pho_organism(new_mutvar,self.rep_don,new_reppnt)
            don_mut = rand.gauss(0,self.mut_var/2)
            if self.rep_don-don_mut<0:
                offspring.rep_don = 0
            else: 
                offspring.rep_don -= don_mut
            return offspring 
        if self.fat<0:
            return False
        else:
            return True
        
    def consume(nutrient_key):
        '''The organism will gather nutrients based on the given nutrient key
        
        nutrient_key: a list of values representing energy from the system
        '''
        if nutrient_key!=None:
            key_slc = min([len(nutrient_key),len(self.key)])
            nutr_key = nutrient_key[:key_slc]
            self_key = self.key[:key_slc]
            diff = sum([abs(nutr_key[i]-self_key[i]) for i in range(key_slc)])
            self.fat += 1./(diff+0.01)
            
class zoo_organism:
    def __init__(self,mut_var,rep_don,rep_pnt,hrm_pnt):
        '''
        mut_var = The mutation rate for the organism.
        rep_don = The nutrients donated to offspring during preproduction
        rep_pnt = An organism reproduces when fat stores exceded rep_pnt
        hrm_pnt = The point at which poor predation results are harmful
        '''
        self.mut_var = mut_var
        self.rep_don = rep_don
        self.rep_pnt = rep_pnt
        self.hrm_pnt = hrm_pnt
        # Initialize Organism's Key
        # ------------------------
        # The key serves multiple purposes and is the central
        # fitness measure. The key is compared to other keys
        # from the enviroment or other organisms. The difference
        # between the individual values of the compared keys is
        # what determines how much energy is transfered between
        # the entities in volved.
        self.key = [rand.random() for i in range(1+int(rand.random()*9))]
        # Initialize Organism's Fat
        # -------------------------
        # Fat is equivalent to the amount of nutrients the organism
        # has stored up. If an organisms fat drops to, or below
        # zero then it dies. If the fat stores exceded rep_pnt (the
        # reproduction point) then the organism reproduces and
        # passes on its characteristics along with the resulting
        # mutations that come along with the process.
        self.fat = rep_don
    
    def sustain(self):
        '''Performs various function to sustain the organism
        
        nutrient_key: nutrient_key represents nutrients from the system
        
        Returns
        -------
        offspring: a spawn with inhereted traits from self if fat>rep_pnt
        False: if fat<0 (indicates that the organism is dead)
        True: if 0<fat<rep_pnt (indicate the organism has sustained itself)
        '''
        self.fat -= len(self.key)/2
        if self.fat>self.rep_pnt:
            self.fat -= self.rep_don
            key_len = int(round(len(self.key)+rand.gauss(0,self.mut_var)))
            new_key = [self.key[i]+rand.gauss(0,self.mut_var) if i>len(self.key) else rand.random() for i in range(key_len)]
            new_mutvar = self.mut_var+rand.gauss(0,self.mut_var/2)
            new_reppnt = self.rep_pnt+rand.gauss(0,self.mut_var*4)
            offspring = pho_organism(new_mutvar,self.rep_don,new_reppnt)
            don_mut = rand.gauss(0,self.mut_var/2)
            if self.rep_don-don_mut<0:
                offspring.rep_don = 0
            else: 
                offspring.rep_don -= don_mut
            return offspring    
        if self.fat<0:
            return False
        else:
            return True
        
    def predate(self,vict=None):
        '''The organism attemps to leach nutrients from its victim
        
        vict: a phy_organism or zoo_organism
        
        The case where vict=None represents a failed predation attempt.
        This causes the predator to lose fat for having attempted, but
        failed to catch anything.
        
        The case where predation is succesful produces two results:
        1) The vict.key and self.key are sufficiently similar and the
           predator steals nutrients from its victim.
        2) The vict.key and self.key are dissimilar such that the sum of
           their differences is greater than hrm_pnt. In this case the
           nutrients from the predator is not able to digest the victim's
           nutrients. This causes both the predator and the victim to
           lose fat; the victim for having been damaged and the predator
           for having been, in a sense, poisoned.
           
        '''
        if vict!=None:
            key_slc = min([len(self.key),len(vict.key)])
            vict_key = vict.key[:key_slc]
            pred_key = self.key[:key_slc]
            diff = sum([abs(pred_key[i]-vict_key[i]) for i in range(key_slc)])
            diff = diff*(self.hrm_pnt-diff)/abs(self.hrm_pnt-diff)/self.hrm_pnt
            fat_cut = 1./(diff+0.005)
            if fat_cut>vict.fat:
                self.fat += vict.fat
                vict.fat -= fat_cut
            else:
                self.fat += fat_cut
                vict.fat -= abs(fat_cut)
        else:
            self.fat -= len(self.key)
```
