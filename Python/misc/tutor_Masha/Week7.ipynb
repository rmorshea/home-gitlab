###ODE exercizes

```python
def deriv(crnt_theta,t):
    '''returns the first and second derivatives of the simple harmonic oscilator diff eq'''
    g = 9.8 #m/s^2
    ell = 0.5 #m
    return np.array([crnt_theta[1],(g/ell)**0.5*np.sin(crnt_theta[0])])
```

```python
t=np.linspace(0,6,500)

syst_init=[]
theta=[]
for i in range(0,5):
    syst_init=np.array([60*i*np.pi/180,0.2])
    theta.append(odeint(deriv,syst_init,t))
#syst_init=(initial angle, initial velocity)
Theta=[]
for i in range(0,5):
    Theta.append(zip(*theta[i]))
```

```python
plt.figure(figsize=(10,10))
for i in range(0,5):
    plt.plot(Theta[i][0],Theta[i][1])
    plt.xlabel('theta');plt.ylabel('omega')
plt.show()

plt.figure(figsize=(10,10))
for i in range(0,5):
    plt.plot(t,Theta[i][1])
    plt.xlabel('t');plt.ylabel('omega')
plt.show()

plt.figure(figsize=(10,10))
for i in range(0,5):
    plt.plot(t,Theta[i][0])
    plt.xlabel('t');plt.ylabel('theta')
plt.show()
```
