```python
def run_syst(length,pop_size,gens,Smut_fact=0,Smut_rate=0,osc_per=None,osc_mag=None,osc_offst=None):
    '''length = characteristics list length
    pop_size = population size
    gens = number of generations (or time steps)
    Smut_fact = the mutations factor for the system. larger mutation
                factors cause the system to have more drastic mutations.
                Only functional if osc_per = "None", as an entry here
                would give the system a non-constant mutation factor.
                (default to no system mutations)
    osc_per = period of the system's mutation factor oscilations.
              if "None", then the system does not mutate or does so
              at a constant rate.
    osc_mag = magnitude of the system's mutation factor oscilations
    osc_offst = the offset for the system's mutation factor oscilations.
                in a function of the form Asin(x)+B, osc_offst = B.
    
    
    runs the system through a specified number of time steps. after
    each time step the system's requirements for fitness can change
    and three organisms will have competed with eachother to reproduce.
    The organisms must then also change in order to accomadate the
    changing conditions so that they can increase their fitness. the
    rate at which the system requirements, and the organisms themselves
    mutate changes over time. The system's mutation rate (when not
    constant) will oscilate with a specified period. The organisms'
    mutation rate is not specified and will evolve in order to increase
    their fitness.'''

        # desirable characteristic
    Y=np.array([0 for i in range(length)])

        # initialize a population:
    Pop=np.array([[20*random.random() for i in range(length)] for j in range(pop_size)])
    P_rates=np.array([random.random() for i in range(pop_size)])
    P_facts=np.array([2*random.random()+0.01 for i in range(pop_size)])
    
        # Smut_fact=0.27*np.cos(float(t)/osc_per)+0.75
    
    Omut_fact=0
    
    if osc_per==None:
        def Smut_func(t,osc_mag,osc_per,osc_offst):
            return Smut_rate
    else:
        def Smut_func(t,osc_mag,osc_per,osc_offst):
            return osc_mag*np.cos(float(t)/osc_per)+osc_offst
    
    def rand_muts(rate):
        if random.random()<rate:
            return 1
        else:
            return 0
    
    def Omut_fact(fact,rate):
        a=3
        while a==0:
            a = fact+random.gauss(0,fact)*rand_muts(rate)
        return a
    
    Survi_facts=[]
    Omut_rates=[]
    Omut_facts=[]
    Smut_rates=[]
    for t in range(gens):
        
        rand=random.random()
        rand_int=random.randint(0,length-1)
        
        Smut_rate=Smut_func(t,osc_mag,osc_per,osc_offst)
        Smut=random.gauss(mu=0,sigma=Smut_fact)
        
            # ind_list represents a list of the three random indicies in the list
            # of orgnisms (that list being Pop)
        randa=randint(0,pop_size-1)
        randb=randint(0,pop_size-1)
        randc=randint(0,pop_size-1)
        while randa==randb:
            randb=randint(0,pop_size-1)
        while randc==randb or randc==randa:
            randc=randint(0,pop_size-1)
        ind_list=np.array([randa,randb,randc])
    
            # A,B,C represent the organisms chosen by each of the indicies in ind_list
            # as such org_list is then a list of those three organisms
        A=Pop[randa]
        B=Pop[randb]
        C=Pop[randc]
        org_list=np.array([A,B,C])
    
            # x_fit is the fitness factors for organism x
            # the fitness factors are made for A,B,C
        A_fit=sum(np.array([np.abs(i-j) for i, j in zip(A, Y)]))
        B_fit=sum(np.array([np.abs(i-j) for i, j in zip(B, Y)]))
        C_fit=sum(np.array([np.abs(i-j) for i, j in zip(C, Y)]))
        fit_list=[A_fit,B_fit,C_fit]

        min_fit=max(fit_list)
        kill=fit_list.index(min_fit)
        live=np.array([i for i in range(3) if i!=kill])
        
        chance=[(True,i) if rand<fit_list[i]/float(min_fit) else (False,i) for i in live]
        
        if chance[0][0]==True and chance[1][0]==True:
            if rand>0.5:
                live=[kill,chance[0][1]] #removable line
                kill=chance[1][1]
            if rand<0.5:
                live=[kill,chance[1][1]] #removeable line
                kill=chance[0][1]
        else:
            for i in range(2):
                if chance[i][0]==True:
                    kill=chance[i][1]
            live=[i for i in range(3) if i!=kill]
        
        live1=Pop[ind_list[live[0]]]
        live2=Pop[ind_list[live[1]]]
        
        kill_ind=ind_list[kill]
        live1_ind=ind_list[live[0]]
        live2_ind=ind_list[live[1]]
        
        Omut_rate1=P_rates[live1_ind]
        Omut_rate2=P_rates[live2_ind]
        Omut1=P_facts[live1_ind]*random.random()
        Omut2=P_facts[live2_ind]*random.random()
        
        avmut_rate=np.average((abs(Omut_rate1),abs(Omut_rate2)))
        
        Pop[kill_ind]=[random.choice((live1[i]+Omut1*rand_muts(P_rates[live1_ind]),live2[i]+Omut2*rand_muts(P_rates[live2_ind]))) for i in range(length)]
        P_rates[kill_ind]=avmut_rate+random.gauss(0,random.choice([P_facts[live1_ind],P_facts[live2_ind]]))
        P_facts[kill_ind]=Omut_fact(P_facts[live1_ind],random.choice([P_facts[live1_ind],P_facts[live2_ind]]))
        
        for i in range(length-1):
            rand=random.random()
            if rand<Smut_rate:
                Y[i]=Y[i]+Smut
            
        Survi_facts.append(sum(np.array([sum(np.array([np.abs(i-j) for i, j in zip(X, Y)])) for X in Pop])))
        Omut_rates.append(P_rates[kill_ind])
        Omut_facts.append(P_facts[kill_ind])
        Smut_rates.append(abs(Smut_rate))
        
    print 'Finnished'
    
    #----------------------------
    
    return range(gens),Survi_facts,Omut_rates,Omut_facts,Smut_rates
```

```python
def run_syst(length,pop_size,gens,Smut_fact=0,Smut_rate=0,osc_per=None,osc_mag=None,osc_offst=None):
    '''length = characteristics list length
    pop_size = population size
    gens = number of generations (or time steps)
    Smut_fact = the mutations factor for the system. larger mutation
                factors cause the system to have more drastic mutations.
                Only functional if osc_per = "None", as an entry here
                would give the system a non-constant mutation factor.
                (default to no system mutations)
    osc_per = period of the system's mutation factor oscilations.
              if "None", then the system does not mutate or does so
              at a constant rate.
    osc_mag = magnitude of the system's mutation factor oscilations
    osc_offst = the offset for the system's mutation factor oscilations.
                in a function of the form Asin(x)+B, osc_offst = B.
    
    
    runs the system through a specified number of time steps. after
    each time step the system's requirements for fitness can change
    and three organisms will have competed with eachother to reproduce.
    The organisms must then also change in order to accomadate the
    changing conditions so that they can increase their fitness. the
    rate at which the system requirements, and the organisms themselves
    mutate changes over time. The system's mutation rate (when not
    constant) will oscilate with a specified period. The organisms'
    mutation rate is not specified and will evolve in order to increase
    their fitness.'''

        # desirable characteristic
    Y=np.array([1 for i in range(length)])

        # initialize a population:
    Pop=np.array([[2*random.random() for i in range(length)] for j in range(pop_size)])
    P_rates=np.array([random.random() for i in range(pop_size)])
    P_facts=np.array([random.random()+0.01 for i in range(pop_size)])
    
        # Smut_fact=0.27*np.cos(float(t)/osc_per)+0.75
    
    Omut_fact=0
    
    if osc_per==None:
        def Smut_func(t,osc_mag,osc_per,osc_offst):
            return Smut_rate
    
    def mut_TF(rate):
        if random.random()<rate:
            return 1
        else:
            return 0
    
    def Omut_fact(fact,rate):
        a=3
        while a==0:
            a = fact+random.gauss(0,fact)*rand_muts(rate)
        return a
    
    Survi_facts=[]
    O_rates=[]
    O_facts=[]
    S_rates=[]
    for t in range(gens):
        
        rand=random.random()
        rand_int=random.randint(0,length-1)
        
            # ind_list represents a list of the three random indicies in the list
            # of orgnisms (that list being Pop)
        randa=randint(0,pop_size-1)
        randb=randint(0,pop_size-1)
        randc=randint(0,pop_size-1)
        while randa==randb:
            randb=randint(0,pop_size-1)
        while randc==randb or randc==randa:
            randc=randint(0,pop_size-1)
        i_list=np.array([randa,randb,randc])
    
            # A,B,C represent the organisms chosen by each of the indicies in ind_list
            # as such org_list is then a list of those three organisms
        A=Pop[randa]
        B=Pop[randb]
        C=Pop[randc]
        org_list=np.array([A,B,C])
    
            # x_fit is the fitness factors for organism x
            # the fitness factors are made for A,B,C
        A_fit=sum(np.array([np.abs(i-j) for i, j in zip(A, Y)]))
        B_fit=sum(np.array([np.abs(i-j) for i, j in zip(B, Y)]))
        C_fit=sum(np.array([np.abs(i-j) for i, j in zip(C, Y)]))
        fit_list=[A_fit,B_fit,C_fit]

        min_fit=max(fit_list)
        kill=fit_list.index(min_fit)
        live=np.array([i for i in range(3) if i!=kill])
        
        chance=[(True,i) if rand<fit_list[i]/float(min_fit) else (False,i) for i in live]
        
        if chance[0][0]==True and chance[1][0]==True:
            if rand>0.5:
                live=[kill,chance[0][1]] #removable line
                kill=chance[1][1]
            if rand<0.5:
                live=[kill,chance[1][1]] #removeable line
                kill=chance[0][1]
        else:
            for i in range(2):
                if chance[i][0]==True:
                    kill=chance[i][1]
            live=[i for i in range(3) if i!=kill]
        
        live1=Pop[i_list[live[0]]]
        live2=Pop[i_list[live[1]]]
        
        kill_i=i_list[kill]
        live1_i=i_list[live[0]]
        live2_i=i_list[live[1]]

        O1_rate=P_rates[live1_i]
        O2_rate=P_rates[live2_i]
        O1_fact=P_facts[live1_i]
        O2_fact=P_facts[live2_i]
        Omut1=P_facts[live2_i]*random.random()
        Omut2=P_facts[live2_i]*random.random()        
        
        Pop[kill_i]=[random.choice((live1[i]+Omut1*mut_TF(O1_rate),live2[i]+Omut2*mut_TF(O2_rate))) for i in range(length)]
        
        a1=mut_TF(O1_rate)
        a2=mut_TF(O2_rate)
        P_facts[kill_i]=random.choice((O1_fact,O2_fact))+random.gauss(0.01,random.choice((O1_fact,O2_fact)))*np.average((a1,a2))*random.choice((O1_fact,O2_fact))
        x=random.choice((O1_rate,O2_rate))+random.random()*random.choice((a1,a2))*random.choice((O1_fact,O2_fact))
        if x>1:
            x=0.99
        if x<0:
            x=0.01
        P_rates[kill_i]=x
        
        O_rates.append(np.average(P_rates))
        O_facts.append(np.average(P_facts))
        Survi_facts.append(sum(sum(np.array([np.abs(i-j) for i, j in zip(X, Y)])) for X in Pop))
        
    print 'Finnished'
    
    #----------------------------
    
    return np.array(range(gens)),np.array(Survi_facts),np.array(O_rates),abs(np.array(O_facts)),np.array(S_rates)
```
