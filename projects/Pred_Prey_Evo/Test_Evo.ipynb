```python
def mut_TF(rate):
    '''determines whether the system will mutate
    with a 0 for False and 1 for True depending on
    the mutation rate input.'''
    if random.random()<rate:
        return 1
    else:
        return 0
    
def Omut_fact(fact,rate):
    a=0
    while a==0:
        a = fact+random.gauss(0,fact)*rand_muts(rate)
    return a
    
def S_Rvar(t,Y,S_fact,osc_mag,osc_per,osc_offst):
    S_rate=osc_mag*np.cos(float(t)/osc_per)+osc_offst
    Y=[i+random.gauss(0,S_fact)*mut_TF(S_rate) for i in Y]
    if S_rate<0:
        S_rate=0
    if S_rate>1:
        S_rate=1
    return S_rate,Y
    
def S_FRvar(t,Y,S_fact,osc_mag,osc_per,osc_offst):
    S_rate=osc_mag*np.cos(float(t)/osc_per)+osc_offst
    
    if S_rate>1:
        S_rate=1
    if S_rate<0:
        S_rate=0
    
    S_fact+=random.gauss(0.01,S_fact)*mut_TF(S_rate/2)
    Y=[(i+random.gauss(0,S_fact)*mut_TF(S_rate)) for i in Y]

    return S_rate,S_fact,Y
```
